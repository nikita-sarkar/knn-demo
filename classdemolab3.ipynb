```python
import numpy as np
import pandas as pd # use for reading and processing the data
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
bcData = pd.read_csv('wisc_bc_data.csv')
```

```python
bcData['diagnosis'] = bcData.diagnosis.astype('category')
```

```python
bcData.describe().transpose()
```

```python
bcData.groupby(['diagnosis']).count()
```

```python
bcData = bcData.drop(labels=['id'], axis = 1)
```

```python
bcData.head()
```

```python
X = bcData.drop(labels=['diagnosis'], axis = 1)
y = bcData['diagnosis']
print(X.head())
print(y.head())
```

```python
zscorenormalization = lambda x : (x-x.mean())/x.std()
XNormalized = X.apply(zscorenormalization)
XNormalized.describe()
```

```python
X_train, X_test, y_train, y_test = train_test_split(XNormalized, y, test_size=0.3, random_state=1)
```

```python
X_train
```

```python
X_test
```

```python
KNN = KNeighborsClassifier(n_neighbors = 5, weights='distance')
```

```python
KNN.fit(X_train, y_train)
```

```python
predicted_labels = KNN.predict(X_test)
KNN.score(X_test, y_test)
```
